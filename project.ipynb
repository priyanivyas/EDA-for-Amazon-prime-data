```python
# importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading dataset
credits = pd.read_csv("/content/credits.csv")
titles = pd.read_csv("/content/titles.csv")
```

#Dataset Description

This data set was created to list all shows available on Amazon Prime streaming, and analyze the data to find interesting facts. This dataset has data available in the United States.

This dataset has two files containing the titles (titles.csv) and the cast (credits.csv) for the title.

titles.csv contains -

id: The title ID on JustWatch.

title: The name of the title.

show_type: TV show or movie.

description: A brief description.

release_year: The release year.

age_certification: The age certification.

runtime: The length of the episode (SHOW) or movie.

genres: A list of genres.

production_countries: A list of countries that produced the title.

seasons: Number of seasons if it's a SHOW.

imdb_id: The title ID on IMDB.

imdb_score: Score on IMDB.

imdb_votes: Votes on IMDB.

tmdb_popularity: Popularity on TMDB.

tmdb_score: Score on TMDB.


credits.csv contains -

person_ID: The person ID on JustWatch.

id: The title ID on JustWatch.

name: The actor or director's name.

character_name: The character name.

role: ACTOR or DIRECTOR

```python
#merge both the data based on common column id
data = pd.merge(credits, titles, on="id")
data.to_csv("merged_output.csv", index=False)
```

**Box plot -**Box plot are used to visualize the data and identify the outliers in data

```python
# box plot for imdb_score
plt.figure(figsize=(8,6))
sns.boxplot(data=data,y='imdb_score')
plt.title('box plot for imdb_score',color='brown')
plt.show()
```

**Insights**

*   The horizontal line within the box represents the median of imdb_score.
*   points plotted beyond the upper and lower are considerd as outliers.

**Histogram -** Histogram represents the distribution of numerical variable.

**Bar plot -** Bar plot represents the categorical data with rectangular bars

#Bivariate Analysis

The Analysis for two variables.

**Sactter Plot -**Scatter values uses dots to represents the values oftwo numerical variables. It shows the relationship between variables.

```python
#scatter plot between imdb_votes and imdb_score
sns.scatterplot(data=data,x='imdb_votes',y='imdb_score',color='black')
plt.title('scatter plot of imdb_votes and imdb_score',color='brown')
plt.show()
```

**Insights -**

*   imdb_score and imdb_votes are positively correlated.




**Line plot -** line plot connects data points with a line to show trends over time across continuous variable.

```python
#line plot for runtime and release_year
plt.figure(figsize=(8,6))
sns.lineplot(data=data,y='runtime',x='release_year')
plt.title('line plot of runtime v/s release year',color='brown')
plt.show()
```

**Insights -**

*   There is slight downward trend in average runtime of movies and TV shows over the years. This indicates that movies and TV shows are getting shorter.
*   This decrease in runtime is because of shorter attention span of viewers, production cost.



**Pie Chart -** Pie chart is used to visualize the proportion of different categories within a whole dataset.

```python
#piechart for type(movies/TV shows)
plt.figure(figsize=(8,6))
plt.pie(data['type'].value_counts(),labels=data['type'].value_counts().index,autopct='%.2f%%')
plt.title('pie char of type(movies and tv shows)',color='brown')
plt.show()
```

**Insights -**


*   movies has higher percentage then shows.
*   Amazon prime focuses on offering a vast collection of movies over TV shows

#Multivariate Analysis

The Analysis for more than two variables.

**Pair plot -**pair plot visualizes the pairwise relationship between variables. It uses scatter plot for relationships and histogram or density plots for individual distributions.

```python
#pair plot
sns.pairplot(data)
plt.show()
```

**Heatmap -** A heatmap uses color intensities to represents the strngth of relationships between numeric variables in a dataset.

```python
num_col=data.select_dtypes(include=np.number)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(num_col.corr(),annot=True,cmap='coolwarm')
plt.show()
```

**Insights -**

*   RED means they are positively correlated.
*   white means they are not correlated.
*   blue means they are negatively correlated.
